```python
import pycountry_convert as pc
def get_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        return continent_code
    except KeyError:
        return None

data2['continent'] = data2['country'].apply(get_continent)
```

#### Checking for Outliers in Commodity Loss Percentages

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=ea_data, x='commodity', y='loss_percentage', hue='commodity', palette='Paired', dodge=False, legend=False)
plt.title('Loss percentages by Commodity')
plt.xticks(rotation=90)
plt.xlabel('Commodity')
plt.ylabel('Loss Percentage')
plt.savefig('boxplot001.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
outlier_list = ['Maize (corn)', 'Millet']
#['Wheat', 'Maize (corn)', 'Rice', 'Sorghum', 'Millet', 'Barley', 'Mangoes, guavas and mangosteens', 'Potatoes', 'Tomatoes', 'Dairy products n.e.c.', 'Beans, dry', 'Sunflower seed', 'Sweet potatoes']
df_outlier_list = ea_data[ea_data['commodity'].isin(outlier_list)]

#IQR for each commodity
outliers_summary = {}
for commodity in outlier_list:
    subset = df_outlier_list[df_outlier_list['commodity'] == commodity]
    Q1 = subset['loss_percentage'].quantile(0.25)
    Q3 = subset['loss_percentage'].quantile(0.75)
    IQR = Q3 - Q1
    
# defining bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
# identifying outliers
    outliers = subset[(subset['loss_percentage'] < lower_bound) | (subset['loss_percentage'] > upper_bound)]
    outliers_summary[commodity] = outliers

    print(f"Commodity: {commodity}")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
    print(f"Number of outliers: {len(outliers)}")
    print(outliers.head(60), "\n")
```

```python
Outlier observations
1. Wheat
In 2013, four outlier loss percentages were recorded in Kenya and Rwanda and in 2010, one outlier loss percentage value was recorded in Kenya.
2. Maize (corn)
```

```python
import geopandas as gpd
world = gpd.read_file("countries/ne_110m_admin_0_countries.shp")
```

```python
ea_data_grouped = ea_data.groupby('country')['loss_percentage'].mean().sort_values(ascending=False).reset_index(name='average_loss_percentage')
ea_data_grouped
```

```python
ea_data1 = ea_data_grouped.merge(world, left_on='country', right_on='SOVEREIGNT', how='left')
ea_data1
```

```python
geo_ea_data = gpd.GeoDataFrame(ea_data1, geometry='geometry')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
geo_ea_data.plot(
    column='average_loss_percentage', 
    cmap='OrRd', 
    legend=True, 
    legend_kwds={
        'label': "Mean Loss Percentage",
        'orientation': "vertical",
        'shrink': 0.5,            
        'pad': 0.02,              
    },
    ax=ax,
    edgecolor='black'
)

ax.set_title('Average Wastage of Foods from East Africa (2000-2022)', fontsize=12, fontweight='bold')
ax.axis('off')
plt.savefig('eamap001.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
# loss percentage by commodity
mean_loss_commodity = ea_data.groupby('commodity')['loss_percentage'].mean().sort_values(ascending=False).reset_index(name='average_loss_percentage')
print(mean_loss_commodity.head())
```

```python
#top 10 commodities with the highest mean loss percentages
plt.figure(figsize=(10, 6))
sns.barplot(data=mean_loss_commodity.head(10), x='average_loss_percentage', y='commodity', hue='commodity', palette='OrRd', dodge=False, legend=False)
plt.title('Top 10 Commodities by Mean Loss Percentage in East African Countries', fontsize=14)
plt.xlabel('Mean Loss Percentage (%)', fontsize=12)
plt.ylabel('Commodity', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#less 10 commodities with the highest mean loss percentages
plt.figure(figsize=(10, 6))
sns.barplot(data=mean_loss_commodity.tail(10), x='average_loss_percentage', y='commodity', hue='commodity', palette='OrRd', dodge=False, legend=False)
plt.title('Least Wasted Commodities in East African Countries', fontsize=14)
plt.xlabel('Mean Loss Percentage (%)', fontsize=12)
plt.ylabel('Commodity', fontsize=12)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
